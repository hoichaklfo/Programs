```python
import random
```

```python
def choose_word():
    with open('sowpods.txt', 'r') as f:
        words = list(f)
    return random.choice(words).upper().strip()
```

```python
def check_word(the_word):
    guess_count = 0
    wrong_count = 0
    done = [True for letter in the_word]
    letter_count = {}
    for letter in the_word:
        letter_count[letter] = 0

    print("_ " * len(the_word))
    running = True
    while running:
        if wrong_count >= 6:
            print("Game is over. ")
            break
        if True not in done:
            print("\nYou got it. ")
            break
        check_letter = True
        while check_letter:
            letter = input("\nGuess a letter: ").upper()
            if letter in the_word:
                letter_count[letter] += 1
                guess_count += 1
                if letter_count[letter] > 1:
                    print("You have guessed this letter, try another letter. ")
                else:
                    check_letter = False
            else:
                check_letter = False
                guess_count += 1
                wrong_count += 1
        for index in range(len(the_word)):
            if the_word[index] == letter or done[index] is False:
                if done[index] is True:
                    print(letter, end=" ")
                    done[index] = False
                elif done[index] is False:
                    print(the_word[index], end=" ")
            else:
                print("_ ", end="")
        if letter not in the_word:
            print("\n\nIncorrect. You have " + str(6 - wrong_count) + " wrong guess(s) left. ")
```

```python
def play():
    new_game = True
    print("You are playing Hangman. ")
    while new_game:
        check_word(choose_word())
        ans = input("Do you want to start a new game? ")
        while True:
            if ans.lower().strip() == "yes":
                break
            elif ans.lower().strip() == "no":
                print("Bye. ")
                new_game = False
                break
            else:
                print("Invalid. ")
                new_game = False
```

```python
play()
```
